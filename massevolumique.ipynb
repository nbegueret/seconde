# Une mesure n'est jamais parfaite, même si elle est réalisée avec le plus grand soin. Il existe toujours des erreurs de mesures.

## Pour obtenir une valeur fiable, il est possible de faire plusieurs mesures et de déterminer l'incertitude u qui en découle. 
## On  peut alors écrire le résultat sous la forme :

# Cela signifie que lorsqu'on note le résultat, on ne garde pas tous les chiffres. 


# Plus l'incertitude u est grande, moins la mesure est fiable. 

<hr/>

## Tous les groupes vont faire leurs mesures et venir noter leurs résultats au tableau. 
## Le programme ci-dessous permet de rentrer les résultats des différents groupes, calcule la valeur moyenne de la masse volumique et l'incertitude u (masse volumique). Si on change de statégie, on relance le programme.

```python
import numpy as np
from math import sqrt

nb_groupe = int(input("Entrer le nombre de groupes dans la classe :  "))

V_liquide = int(input("Entrer en mL le volume de liquide mesuré par chaque groupe :  "))
i=0
masse_volumique=[]
for i in range (nb_groupe) :
    print("Groupe",i)
    masse =float(input("Entrer, en g, la masse mesurée par le groupe : "))
    masse_volumique.append( masse/V_liquide)
somme =sum(masse_volumique)
moyenne =somme/nb_groupe
print("La masse volumique moyenne est", moyenne,"g/ml")
racine_carre= sqrt(nb_groupe)
ecart_type=np.std(masse_volumique)
u = ecart_type/racine_carre
print ("L'incertitude sur la valeur de la masse volumique est", u , "g/mL")
```
